```python
import numpy as np
import pandas as pd
import os
import glob

# use glob to get all the csv files in the folder
path = os.getcwd()
csv_files = glob.glob(os.path.join(path, "*.csv"))
df_list = []
# loop over the list of csv files
for f in csv_files:
    # read the csv file
    df_list.append(pd.read_csv(f))
```

```python
df_list[0].head()
```

```python
#df_list[1].head()
```

```python
#Kiểm tra số chiều của các dataframe
for i in range(len(df_list)):
    print(f"Dataframe {i+1}: {df_list[i].shape}")
```

```python
#kiểm tra các cột bị thiếu
all_columns1 = set().union(*[df.columns for df in df_list])
drop_columns1 = set()
for i,df in enumerate(df_list):
    missing_columns1 = all_columns1 - set(df.columns)
    drop_columns1 |= missing_columns1
    if missing_columns1:
        print(f"\nDataframe {i+1} thiếu các cột: {missing_columns1}")
```

```python
# Đổi tên cột pagesNumber cho đồng bộ dữ liệu
for df in df_list:
    columns = df.columns.tolist()
    for col in columns:
        if 'pagesnumber' in col.lower():
            df = df.rename(columns={col: 'PagesNumber'}, inplace = True)
```

```python
# Kiểm tra lại các cột bị thiếu và thực hiện xóa trong các df còn lại
all_columns = set().union(*[df.columns for df in df_list])
drop_columns = set()
for i,df in enumerate(df_list):
    missing_columns = all_columns - set(df.columns)
    drop_columns |= missing_columns
    if missing_columns:
        print(f"\nDataframe {i+1} thiếu các cột: {missing_columns}")
```

```python
for df in df_list:
    for column in drop_columns:
        if column in df.columns:
            df.drop(column, axis=1, inplace=True)
```

```python
#Kiểm tra lại số chiều sau khi xóa các cột bị thiếu
for i in range(len(df_list)):
    print(f"Dataframe {i+1}: {df_list[i].shape}")
```

```python
# Đồng bộ thứ tự các cột theo dataframe 6
list_columns = df_list[5].columns
for df in df_list:
    df = df.reindex(columns=list_columns)
    #df = df[list_columns]
#df_list[0].head()
```

```python
# Gộp các dataframe thành một
new_df = pd.concat(df_list, axis=0, ignore_index=True)
new_df.head()
```

```python
# Nếu tháng > thì đổi tên 2 cột ngày, tháng, lỗi quan sát được
for df in df_list:
    if df['PublishMonth'].max()>12:
        df.rename(columns={'PublishMonth':'PublishDay','PublishDay': 'PublishMonth'}, inplace = True)

#in ra df đầu để quan sát
df_list[0].head()
```

```python
# Gộp các cột ngày, tháng, năm làm một cột mới PublishDate
date_cols = ['PublishYear', 'PublishMonth', 'PublishDay']
new_df['PublishDate'] = new_df[date_cols].applymap(str).apply('-'.join, axis=1)
new_df['PublishDate'] = pd.to_datetime(new_df['PublishDate'],errors='coerce')

invalid_rows = new_df[new_df['PublishDate'].isnull()].sort_values(by = 'PublishYear')
i=0
if not invalid_rows.empty:
    print("Invalid rows:")
    print(invalid_rows.loc[:,['ISBN', 'PublishDay', 'PublishMonth','PublishYear']])
else: print('Các giá trị đều hợp lệ')
```
